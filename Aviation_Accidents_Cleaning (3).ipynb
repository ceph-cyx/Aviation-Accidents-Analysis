### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python

df1['SeriousFatalInjury'] = df1['InjuryLevel'].isin(['Fatal', 'Serious']).astype(int)

df1['SeriousFatalInjury'].value_counts()
df1.head()
```

```python
meter = ['Fatalities', 'InjuryLevel', 'Over2250', 'Over5700', 'ScheduledCommercial']
df1[meter].isnull().sum
```

```python
#clean injury level
df1['InjuryLevel']=df1['InjuryLevel'].fillna('None')
df1['InjuryLevel'].value_counts(dropna=False)
```

```python
#clean fatalities
fatal_median = df1.loc[df1['InjuryLevel'] == 'Fatal', 'Fatalities'].median()

#gather fatalities
def impute_fatalities(row):
    if pd.notnull(row['Fatalities']):
        return row['Fatalities']
    elif row['InjuryLevel'] in ['None', 'Minor']:
        return 0
    elif row['InjuryLevel'] == 'Fatal':
        return fatal_median
    else:
        return fatal_median
#apply by row
df1['Fatalities'] = df1.apply(impute_fatalities, axis=1)
#null confirmation
df1['Fatalities'].isnull().sum()
```

```python
#clean occcats
#Occats are standard occurance codes isiued by the icao which may impy destruction
df1['OccCats'].isnull().sum()
```

```python
#clean risk, UNK for unknown category which may be irrelevant
df1['Risk'] = df1['Risk'].fillna('UNK')
df1['Risk'].isnull().sum()
```

#### occarance codes in occats will be used to estimate aircraft damage
CFIT - Controlled Flight Into Terrain 
LOC-I - Loss of Control In-flight 
F-POST - Post impact fire 
F-NI - Fire/Smoke Non-Impact
MAC - Mid Air Collision 


```python

codes = {'CFIT', 'LOC-I', 'F-POST', 'F-NI', 'MAC'}

#rows risk asd fatal estimate to amount of damage
def destroyed(row):
    risk_destroyed = row['Risk'] in codes
    fatal_destroyed = (
        row['InjuryLevel'] == 'Fatal' and 
        row['Risk'] != 'RS'
    )
    return int(risk_destroyed or fatal_destroyed)

df1['Destroyed'] = df1.apply(destroyed, axis=1)

df1['Destroyed'].value_counts()
```
